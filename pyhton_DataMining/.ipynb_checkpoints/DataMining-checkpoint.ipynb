```python
completedmath = StudPerfo["math score"][(StudPerfo['test preparation course'] == 'completed')]
samplemath1=completedmath.sample(n=200, replace="False")
```

```python
nonemath = StudPerfo["math score"][(StudPerfo['test preparation course'] == 'none')] 
samplemath2=nonemath.sample(n=200, replace="False")
```

```python
sta.variance(samplemath1)
```

```python
sta.variance(samplemath2)
```

```python
test_wert1, p_wert1 = sc.shapiro(samplemath1)
print("Für die erste Stichprobe")
print("Test-Wert: ",test_wert1)
print("P-Wert:    ",p_wert1)
if p_wert1 > 0.05:
    print("Erste Stichprobe ist normalverteilt.")
else:
        print("Erste Stichprobe ist nicht normalverteilt.")
print("---")
test_wert2, p_wert2 = sc.shapiro(samplemath2)
print("Für die zweite Stichprobe")
print("Test-Wert: ",test_wert2)
print("P-Wert:    ",p_wert2)
if p_wert2 > 0.05:
    print("Zweite Stichprobe ist normalverteilt.")
else:
        print("Zweite Stichprobe ist nicht normalverteilt.")
```

```python
sns.distplot(samplemath1)
plt.title("Histogram of samplemath1")
plt.savefig('sample1_function.jpg', dpi = 100)
```

```python
sns.distplot(samplemath2)
plt.title("Histogram of samplemath2")
plt.savefig('sample2_function.jpg', dpi = 100)
```

```python
alfa = 0.05
t_test_sta1, p_wert3 = sc.ttest_ind(samplemath1, samplemath2, equal_var=False)
print("t-Test-Statistik: ",t_test_sta1)
print("P-Wert:      ",p_wert3)
if p_wert3 > alfa :
    print("Die Nullhypothese ist richtig. Mathematik-Testdurchschnitte von Männern und Frauen sind gleich.")
else:
    print("Die Alternativhypothese kann nicht abgelehnt werden. Es gibt einen deutlichen Unterschied.")
```

```python
mu = 66.089
stand_devia = 15.163
x_values = np.arange(0, 100, 1)
y_values = sc.norm(mu, stand_devia)
plt.figure(figsize=(11,7))
plt.ylabel("Wahrscheinlichkeitsdichte", fontsize='16')
plt.plot(x_values, y_values.pdf(x_values))
plt.axvline(x=66.089, color='red', linewidth=6)
plt.title("Verteilung der Mathematiknoten:  Mean:66.089  SD:15.163", fontsize=16)
plt.savefig("out13.jpg")
```

```python
def draw_z_score(x, cond, mu, sigma, title):
    y = sc.norm.pdf(x, mu, sigma)
    z = x[cond]
    plt.plot(x, y, color="b")
    plt.fill_between(z, 0, sc.norm.pdf(z, mu, sigma), color="red")
    plt.title(title, fontsize='16')
x = np.arange(-3,3,0.001)
z0 = 0.9174257
plt.figure(figsize=(11,7))
draw_z_score(x, x<z0, 0, 1, 'Mathematiknoten(SNV)  (Z-Wert:0.9174257) (Mean:0) (SD:1)')
plt.ylabel("Wahrscheinlichkeitsdichte", fontsize='16')
plt.xlabel("Z-Wert", fontsize='16')
plt.figure(figsize=(11,7))
plt.savefig("ou5t.jpg")
```

```python
sample_female_bachelor = StudPerfo["writing score"][(StudPerfo['test preparation course'] =='completed') & (StudPerfo['gender']=='female')]
sample_female_bachelor=sample_female_bachelor.sample(n=55, replace="False")
```

```python
sample_male_bachelor = StudPerfo["writing score"][(StudPerfo['test preparation course'] =='completed') & (StudPerfo['gender']=='male')]
sample_male_bachelor=sample_male_bachelor.sample(n=55, replace="False")
```

```python
sta.variance(sample_female_bachelor)
```

```python
sta.variance(sample_male_bachelor)
```

```python
test_wert1, p_wert1 = sc.shapiro(sample_female_bachelor)
print("Für die erste Stichprobe")
print("Test-Wert: ",test_wert1)
print("P-Wert:    ",p_wert1)
if p_wert1 > 0.05:
    print("Erste Stichprobe ist normalverteilt.")
else:
        print("Erste Stichprobe ist nicht normalverteilt.")
print("---")
test_wert2, p_wert2 = sc.shapiro(sample_male_bachelor)
print("Für die zweite Stichprobe")
print("Test-Wert: ",test_wert2)
print("P-Wert:    ",p_wert2)
if p_wert2 > 0.05:
    print("Zweite Stichprobe ist normalverteilt.")
else:
        print("Zweite Stichprobe ist nicht normalverteilt.")
```

```python
alfa = 0.05
t_test_sta1, p_wert3 = sc.ttest_ind(sample_female_bachelor, sample_male_bachelor, equal_var=False)
print("t-Test-Statistik: ",t_test_sta1)
print("P-Wert:           ", p_wert3)
if p_wert3 > alfa :
    print("Die Nullhypothese ist richtig. Mathematik-Testdurchschnitte von Männern und Frauen sind gleich.")
else:
        print("Die Alternativhypothese kann nicht abgelehnt werden. Es gibt einen deutlichen Unterschied.")
```
